```
import gpt_2_simple as gpt2
```

```
gpt2.download_gpt2()
```

```
love = '/content/drive/My Drive/Messi/love.txt'
```

```
sess = gpt2.start_tf_sess()

gpt2.finetune(
              sess,
              dataset=love,
              model_name='124M',
              steps = 1000,
              restore_from = 'fresh',
              run_name ='runx',
              print_every=20,
              sample_every=100,
              save_every=500
              )
```

```
gpt2.generate(sess, run_name='runx')
```
